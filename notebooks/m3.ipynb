```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("/home/caden/Programming/m3/data/processed/M3 - imputed (4).csv")
```

```python
US = ["US ENVIRONMENT","US GDP PC", "US BIKE COOL", "US MED AGE", "US GAS PRICE", "US GYM", "LITHIUM", "US SPENDING", "US BIKE SALES"]
US_DATA = data[US] 

UK = ["UK ENVIRONMENT","UK GDP PC", "UK BIKE COOL", "UK MED AGE", "UK GAS PRICE", "UK GYM", "LITHIUM", "UK SPENDING", "UK BIKE SALES"]
UK_DATA = data[UK] 

st_scaler = preprocessing.StandardScaler()

US_DATA = st_scaler.fit_transform(US_DATA)
US_DATA = pd.DataFrame(US_DATA, columns=US)
UK_DATA = st_scaler.fit_transform(UK_DATA)
UK_DATA = pd.DataFrame(UK_DATA, columns=UK)
```

```python
US_X_pred = ["US ENVIRONMENT","US GDP PC", "US BIKE COOL", "US MED AGE", "US GAS PRICE", "US GYM", "LITHIUM", "US SPENDING"]
US_Y_pred = ["US BIKE SALES"]

X_US = US_DATA[US_X_pred]
Y_US = US_DATA[US_Y_pred]

UK_X_pred = ["UK ENVIRONMENT","UK GDP PC", "UK BIKE COOL", "UK MED AGE", "UK GAS PRICE", "UK GYM", "LITHIUM", "UK SPENDING"]
UK_Y_pred = ["UK BIKE SALES"]

X_UK = UK_DATA[UK_X_pred]
Y_UK = UK_DATA[UK_Y_pred]
```

```python
X_UK.head()
```

```python
model = LinearRegression()
model.fit(X_US, Y_US)
us_coef = model.coef_
print(f"Coef for us: ")
print(us_coef)

model = LinearRegression()
model.fit(X_UK, Y_UK)
uk_coef = model.coef_
print(f"Coef for uk: ")
print(uk_coef)
```

```python
import random
```

```python
predictor_array = US_X_pred
for predictor in predictor_array:
    col_to_jitter = X_US[predictor].copy()
    for i in range(len(col_to_jitter)):
        col_to_jitter[i] = random.uniform(col_to_jitter[i] * 0.95, col_to_jitter[i] * 1.05)
    demo_X_US = X_US.copy()
    demo_X_US[predictor] = col_to_jitter
    # model
    model = LinearRegression()
    model.fit(demo_X_US, Y_US)
    coef = model.coef_

    display_arr = []
    for i in range(len(us_coef[0])):
        temp = us_coef[0][i] - coef[0][i]
        display_arr.append(temp/coef[0][i] * 100)

    print(f"Coef for {predictor}: ")
    print(display_arr)
```

```python
predictor_array = UK_X_pred
for predictor in predictor_array:
    col_to_jitter = X_UK[predictor].copy()
    for i in range(len(col_to_jitter)):
        col_to_jitter[i] = random.uniform(col_to_jitter[i] * 0.95, col_to_jitter[i] * 1.05)
    demo_X_UK = X_UK.copy()
    demo_X_UK[predictor] = col_to_jitter

    # model
    model = LinearRegression()
    model.fit(demo_X_UK, Y_UK)
    coef = model.coef_

    display_arr = []
    for i in range(len(us_coef[0])):
        temp = uk_coef[0][i] - coef[0][i]
        display_arr.append(temp/coef[0][i] * 100)
        
    print(f"Coef for {predictor}: ")
    print(display_arr)

    
```
