```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = ''
Y = ''
```

```python
plt.scatter(X, Y)
```

```python
# estimate shape of scatter plot
degree = 2

# include bias forces y-int to be 0
poly = PolynomialFeatures(degree=degree, include_bias=False)

#reshape data to work properly with sklearn
reshaped_X = poly.fit_transform(X.reshape(-1, 1))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(reshaped_X, Y, test_size = 0.2, random_state=1)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
model = LinearRegression(random_state=1)
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, Y_pred)))
r2 = r2_score(Y_test, Y_pred)

print("The model performance for testing set")
print("--------------------------------------")
print(f'RMSE is {rmse}')
print(f'R2 score is {r2}')

intercept = model.intercept_
coef = model.coef_

print(intercept, coef)
```
