```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

url = "./data/raw/caden_wenis.csv"
data = pd.read_csv(url, header=[1])
data.head()
```

```python
df = data[["par_pctile", "tier", "count"]]
df.par_pctile.unique()
```

```python

import seaborn as sns
import numpy as np 
pctiles = df.par_pctile.unique()
tiers = df.tier.unique()
data = np.empty([len(tiers),len(pctiles)])
print(pctiles)
print(tiers)
for i in range(len(tiers)):
  for j in range(len(pctiles)):
    data[i][j] = df.loc[ (df['par_pctile'] == pctiles[j]) & (df['tier'] == tiers[i])]['count']

fig, ax = plt.subplots(figsize=(70,70))
im = ax.imshow(data,cmap='YlOrRd')
ax.set_xticks(np.arange(len(pctiles)), labels=pctiles)
ax.set_yticks(np.arange(len(tiers)), labels=tiers)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
for i in range(len(tiers)):
    for j in range(len(pctiles)):
        text = ax.text(j,i, data[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
# fig.tight_layout()
plt.show()
```
