```python
import numpy as np
```

```python
state_1 = [0.9,0.1]
state_2 = [0.5,0.5]
# unused in this example
state_3 = [1,1,1,1]

initial_system = np.array([.5,.5])

iterations = 10
```

```python
# Brackets here are necessary because you need to concatenate 2D arrays into 2D arrays, not 1D into 2D
state_1 = np.array([state_1])
state_2 = np.array([state_2])
state_3 = np.array([state_3])

system = np.concatenate((state_1,state_2), axis=0)

print(system)
```

```python
current_system = np.dot(initial_system,system)
running = []

for i in range(iterations) :
  probability = np.dot(current_system,system)
  
  running.append(probability)

  current_system = probability

for i in range(len(running)) :
  running[i] = np.array([running[i]])

all_systems = np.concatenate(running, axis=0)


all_systems = np.concatenate((np.array([initial_system]),all_systems), axis=0)
print(all_systems)
```

```python
import matplotlib.pyplot as plt
state_1 = all_systems[:, 0, None]
state_2 = all_systems[:, 1, None]
plt.plot(state_1, label="state_1")
plt.plot(state_2, label="state_2")
plt.legend()
```
