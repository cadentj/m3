```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import normal
from math import ceil
```

```python
# Inputs

# Outputs (tracking)
probability = []
```

```python
# Define function here
def number_of_stops(distance, charge_full, charge_zero, truck_range) :
    return ceil(distance / ( (charge_full - charge_zero) * truck_range))
```

```python
iterations = 1000

for i in range(iterations) :

    # Initialize random variables
    truck_range = normal(300,100)
    charge_full = normal(0.6,0.125)

    # Constant variables
    san_antonio_new_orleans = 533 # miles
    charge_zero = 0.2 # assumption: trucks recharge at 20% to ensure they don't run out of power

    # Function
    result = number_of_stops(san_antonio_new_orleans, charge_full, charge_zero, truck_range)

    probability.append(result)
```

```python
p = np.array(probability)

plt.hist(p)
plt.xlabel("Number of Stops")
plt.ylabel("Freq")

# Add a vertical line at the mean
plt.axvline(p.mean(), color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(p.mean() * 1.5, max_ylim*0.9, 'Mean: {:.2f}'.format(p.mean()))

plt.show()
```

```python
avg_probability = sum(probability)/len(probability)
print(f"Mean: {avg_probability}")
```
